```python
# Import Tools

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from keras.layers import Dense
from keras.models import Sequential
from keras.optimizers import SGD
from keras.callbacks import EarlyStopping
from sklearn.preprocessing import StandardScaler
from keras.utils import to_categorical

# Make plots pretty
sns.set()
```

```python
# Read in Data
df = pd.read_csv('Combo_Cleaned.csv')
df.drop('Unnamed: 0', axis=1, inplace=True)
df.head()
```

```python
# Create our X and y for the model with individual y's for each metallic tier.

# Our X Dropping the Metallic Tier Targets, String Data, and Categorical Year
X = df.drop(['cfc','state','county','platinum','gold','silver','bronze','catastrophic', 'Year',
             'tps'], axis=1).values

# Get shape of predictors
n_cols = X.shape[1]

# Individual y's for Each Target Data
ycat = df.catastrophic.valuestype
ybro = df.bronze.values
ysil = df.silver.values
ygold = df.gold.values
yplat = df.platinum.values

# Create List of Target Values and Names of Target Values
ys=[ycat, ybro, ysil, ygold, yplat]
ys_name = ['Catastrophic','Bronze','Silver','Gold','Platinum']
```

```python
# Define model creation program.

def make_model(input_shape = (n_cols,)):
    model = Sequential()
    model.add(Dense(100, activation='relu', input_shape=input_shape))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    return(model)

# Create Early Stopping monitor
monitor = EarlyStopping(patience=3)
```

```python
make_model()
ss = StandardScaler()
Xs = ss.fit_transform(X)
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(Xs, ysil,validation_split=.3, epochs=5, batch_size=128)
```

```python
# Read in Data
dfp = pd.read_csv('Combo_Percents.csv')
dfp.drop('Unnamed: 0', axis=1, inplace=True)
dfp.head()
```

```python
# Create the X and y for the model with individual y's for each metallic tier.

# Our X Dropping the Metallic Tier Targets, String Data, Categorical Year, and TPS
Xp = dfp.drop(['cfc','state','county','platinum','gold','silver','bronze','catastrophic', 'Year',
               'tps'],axis=1).values

# Individual y's for Each Target Data
ycatp = (dfp.catastrophic.values * 100).astype(int)
ybrop = (dfp.bronze.values *100).astype(int)
ysilp = (dfp.silver.values * 100).astype(int)
ygoldp = (dfp.gold.values * 100).astype(int)
yplatp = (dfp.platinum.values * 100).astype(int)

# Get shape of predictors
n_cols_class = Xp.shape[1]

# Create List of Target Values for Percent Data
ysp=[ycatp, ybrop, ysilp, ygoldp, yplatp]
```

```python
# Define program that converts target percents of TPS to categorical numbers
def convert(target):
    '''Takes array from 0 - 100 and converts it to categories representing batches of 5'''
    for n in range(len(target)):
        if target[n] in np.arange(0,5):
            target[n] = 0
        if target[n] in np.arange(5,10):
            target[n]=1
        if target[n] in np.arange(10,15):
            target[n] = 2
        if target[n] in np.arange(15,20):
            target[n]=3
        if target[n] in np.arange(20,25):
            target[n] = 4
        if target[n] in np.arange(25,30):
            target[n]=5
        if target[n] in np.arange(30,35):
            target[n] = 6
        if target[n] in np.arange(35,40):
            target[n]=7
        if target[n] in np.arange(40,45):
            target[n]=8
        if target[n] in np.arange(45,50):
            target[n] = 9
        if target[n] in np.arange(50,55):
            target[n] = 10
        if target[n] in np.arange(55,60):
            target[n]=11
        if target[n] in np.arange(60,65):
            target[n] = 12
        if target[n] in np.arange(65,70):
            target[n]=13
        if target[n] in np.arange(70,75):
            target[n] = 14
        if target[n] in np.arange(75,80):
            target[n]=15
        if target[n] in np.arange(80,85):
            target[n] =16
        if target[n] in np.arange(85,90):
            target[n]=17
        if target[n] in np.arange(90,95):
            target[n]=18
        if target[n] in np.arange(95,101):
            target[n]=19
```

```python
# Define model creation program.
def make_model_class(predictor, input_shape = (n_cols_class,)):
    model = Sequential()
    model.add(Dense(100, activation='relu', input_shape=input_shape))
    model.add(Dense(100, activation='relu'))
    model.add(Dense(100, activation='relu'))
    model.add(Dense(len(np.unique(predictor)), activation='softmax'))
    return(model)

# Create Early Stopping monitor
monitor = EarlyStopping(patience=3)
```

```python
# Convert Catastrophic Tier to Binned Categories
convert(ycatp)
```

```python
# Test Classification Model on Catastrophic Dataset
test = ycatp/10
make_model_class(predictor=ycatp)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(Xp, test,validation_split=.3, epochs=10, callbacks=[monitor])
```
