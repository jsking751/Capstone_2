```python
# Import Necessary Tools
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Make Plots Pretty
sns.set()
```

```python
# Import Datasets
df16 = pd.read_csv('2016_Cleaned.csv')
df15 = pd.read_csv('2015_Cleaned.csv')
master = pd.read_csv('Combo_Cleaned.csv')

df16.drop('Unnamed: 0', axis=1, inplace=True)
df15.drop('Unnamed: 0', axis=1, inplace=True)
master.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
# Compute Bins for Each Dataframe
#2015
data15 = len(df15.tps)
data15 = len(df15.tps)
bins15= int(np.sqrt(data15))
#2016
data16 = len(df16.tps)
data16 = len(df16.tps)
bins16= int(np.sqrt(data16))

print('Histogram Bins: ','\n','2015: ',bins15, '\n', '2016: ', bins16)
```

TPS is the chosen distribution check as it is the total plan selections for each observation.  As such, it is most reflective of all datapoints within each observation.

```python
# Plotting Star Rating Histogram
fig, axes = plt.subplots(1,2, figsize=(12,5))

_ = df15.tps.plot(kind='hist', bins=bins15, ax=axes[0], title = '2015 TPS Distribution',)
_ = axes[0].set_ylabel('Count')
_ = axes[0].set_xlabel('Total Plan Selections')

_ = df16.tps.plot(kind='hist', bins=bins16, ax=axes[1], title = '2016 TPS Distribution')
_ = axes[1].set_ylabel('Count')
_ = axes[1].set_xlabel('Total Plan Selections')

plt.show()
```

Due to a few heavily populated counties it appears the data is heavily right tailed.  Let's zoom into the data between 0 and 10,000.

```python
# Use List Comprehension to Filter Out TPS Greater Than 50000
reduced15= list(filter(lambda x: x < 10000, df15.tps))
reduced16 = list(filter(lambda x: x < 10000, df16.tps))
```

```python
# Plotting TPS Filtered Histogram
fig, axes = plt.subplots(1,2, figsize=(12,5))

_ = axes[0].hist(reduced15)
_ = axes[0].set_ylabel('Count')
_ = axes[0].set_xlabel('Total Plan Selections')
_ = axes[0].set_title('2015 TPS Distribution (Filtered)')

_ = axes[1].hist(reduced16)
_ = axes[1].set_ylabel('Count')
_ = axes[1].set_xlabel('Total Plan Selections')
_ = axes[1].set_title('2016 TPS Distribution (Filtered)')

plt.show()
```

Data distribution appears to be heavily right tailed, even after zooming in.

Let's take a look at the distribution of Metalic Tiers within the data.

```python
# Set up Tier Order
tier_lvl = ['catastrophic','bronze','silver','gold','platinum']

# Melt 2015 Data to be Pivoted and Rename Columns
tiers15 = pd.melt(df15[['platinum','gold','silver','bronze','catastrophic']])
tiers15.columns = ['tiers', 'count']

# Pivot 2015 Data Reset Index
tiers15 = pd.pivot_table(tiers15, index = 'tiers', values = 'count', aggfunc=np.sum)
tiers15.reset_index(inplace=True)

# Set 2015 Tier Data as Categorical and Sort by Tier Level
tiers15.tiers = pd.Categorical(tiers15.tiers, categories=tier_lvl, ordered=True)
tiers15.sort_values('tiers', inplace=True)

# Melt 2016 Data to be Pivoted and Rename Columns
tiers16 = pd.melt(df16[['platinum','gold','silver','bronze','catastrophic']])
tiers16.columns = ['tiers', 'count']

# Pivot 2016 Data Reset Index
tiers16 = pd.pivot_table(tiers16, index = 'tiers', values = 'count', aggfunc=np.sum)
tiers16.reset_index(inplace=True)

# Set 2016 Tier Data as Categorical and Sort by Tier Level
tiers16.tiers = pd.Categorical(tiers16.tiers, categories=tier_lvl, ordered=True)
tiers16.sort_values('tiers', inplace=True)

# Plot Metalic Tiers Count
fig, axes = plt.subplots(1,2, figsize=(15,5))

_ = sns.barplot(x='tiers', y='count', data=tiers15, color='navy', ax=axes[0])
_ = axes[0].set_xlabel('Metalic Tiers')
_ = axes[0].set_yticklabels(['{:,}'.format(int(x)) for x in axes[0].get_yticks().tolist()])
_ = axes[0].set_ylabel('Count')
_ = axes[0].set_title('2015 Metalic Tier Distributions')

_ = sns.barplot(x='tiers', y='count', data=tiers16, color='navy', ax=axes[1])
_ = axes[1].set_xlabel('Metalic Tiers')
_ = axes[1].set_yticklabels(['{:,}'.format(int(x)) for x in axes[1].get_yticks().tolist()])
_ = axes[1].set_ylabel('Count')
_ = axes[1].set_title('2016 Metalic Tier Distributions')

plt.show()
```

```python
# Plot Pie Charts Of Metalic Tier Distributions
import plotly.plotly as py
import plotly.graph_objs as go

labels = ['Catastrophic','Bronze','Silver','Gold','Platinum']
sizes15 = tiers15['count']
sizes16 = tiers16['count']
colors = ['black', 'darkgoldenrod', 'silver', 'gold', 'whitesmoke']

# Set plotly Dictionaries
fig = {'data':[{'labels':labels, 'values':sizes15, 'type':'pie', 'marker':{'colors':colors},
                'domain':{'x':[0,.5],'y':[0,1]}, 'hoverinfo':'label+percent','textinfo':'perecent','name':'2015'},
              {'labels':labels, 'values':sizes16, 'type':'pie', 'marker':{'colors':colors},
                'domain':{'x':[.55,.99],'y':[0,1]}, 'hoverinfo':'label+percent','textinfo':'percent', 'name':'2016'}],
       'layout': {'title': 'Metallic Tier Pie Charts', 'showlegend': True,"annotations":[{"font": {"size": 20},
                "showarrow": False,"text": "2015","x": 0.21,"y":1.1},
                {"font": {"size": 20},"showarrow": False,"text": "2016","x": 0.8,"y": 1.1}]}}
# Plot
py.iplot(fig, filename='metallic_pie')
```

The metalic tier counts appear to potentially be normally distributed for both 2015 and 2016.

```python
# Create Correlation Matrix to See Noteable Correlaitons
f, ax = plt.subplots(figsize=(15, 15))
corr = df15.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            center=0,square=True, ax=ax)
plt.title('2015 Data Heatmap')
plt.show()
```

```python
# Create Correlation Matrix to See Noteable Correlaitons
f, ax = plt.subplots(figsize=(15, 15))
corr = df16.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            center=0,square=True, ax=ax)
plt.title('2016 Data Heatmap')
plt.show()
```

```python
# Create Correlation Matrix to See Noteable Correlaitons
f, ax = plt.subplots(figsize=(15, 15))
corr = master.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            center=0,square=True, ax=ax)
plt.title('Master Heatmap')
plt.show()
```

```python
# Melt 2015 Age Data
ages= ['0-17','18-25','26-34','35-44','45-54','55-64','65+']
ages15 = pd.melt(df15[ages])

# Convert 2015 Ages to Pivot Table
ages15 = pd.pivot_table(ages15, index = 'variable', values = 'value', aggfunc=np.sum)

# Melt 2016 Age Data
ages16 = pd.melt(df16[ages])

# Convert 2016 Ages to Pivot Table
ages16 = pd.pivot_table(ages16, index = 'variable', values = 'value', aggfunc=np.sum)


# Plot Age Counts
fig, axes = plt.subplots(1,2, figsize=(15,5))

_ = sns.barplot(x=ages15.index, y='value', data=ages15, color='green', ax=axes[0])
_ = axes[0].set_xlabel('Ages')
_ = axes[0].set_ylabel('Count')
_ = axes[0].set_yticklabels(['{:,}'.format(int(x)) for x in axes[0].get_yticks().tolist()])
_ = axes[0].set_title('2015 Age Distributions')

_ = sns.barplot(x=ages16.index, y='value', data=ages16, color='green', ax=axes[1])
_ = axes[1].set_xlabel('Ages')
_ = axes[1].set_ylabel('Count')
_ = axes[1].set_yticklabels(['{:,}'.format(int(x)) for x in axes[1].get_yticks().tolist()])
_ = axes[1].set_title('2016 Age Distributions')

plt.show()
```

```python
# Melt 2015 Income Data
fpl15 = ['fpl_<_100','fpl_<_151','fpl_<_201','fpl_<_251','fpl_<_301','fpl_<_401', 'fpl_>_400','fpl_unknown']
income15 = pd.melt(df15[fpl15])

# Convert 2016 Income Data to Pivot Table
income15 = pd.pivot_table(income15, index = 'variable', values = 'value', aggfunc=np.sum)

# Melt 2016 Income Data
fpl16 = ['fpl_<_100','fpl_<_139','fpl_<_151','fpl_<_201','fpl_<_251','fpl_<_301','fpl_<_401', 'fpl_>_400','fpl_unknown']
income16 = pd.melt(df16[fpl16])

# Convert 2016 Income Data to Pivot Table
income16 = pd.pivot_table(income16, index = 'variable', values = 'value', aggfunc=np.sum)

# Plot 2015 Income Counts
fig, axes = plt.subplots(1,2, figsize=(15,5))

_ = sns.barplot(x=income15.index, y='value', data=income15, color='goldenrod', ax=axes[0])
_ = axes[0].set_xlabel('Income by Federal Poverty Limit Percent')
_ = axes[0].set_ylabel('Count')
_ = axes[0].set_title('2015 Income Distributions')
_ = axes[0].set_xticks([n for n in range(len(fpl15))])
_ = axes[0].set_yticklabels(['{:,}'.format(int(x)) for x in axes[0].get_yticks().tolist()])
_ = axes[0].set_xticklabels(['0-99','100-150','151-200','201-250','251-300','301-400','>400','Unknown'])

_ = sns.barplot(x=income16.index, y='value', data=income16, color='goldenrod', ax=axes[1])
_ = axes[1].set_xlabel('Income by Federal Poverty Limit Percent')
_ = axes[1].set_ylabel('Count')
_ = axes[1].set_title('2016 Income Distributions')
_ = axes[1].set_xticks([n for n in range(len(fpl16))])
_ = axes[1].set_yticklabels(['{:,}'.format(int(x)) for x in axes[1].get_yticks().tolist()])
_ = axes[1].set_xticklabels(['0-99','100-138','139-150','151-200','201-250','251-300','301-400','>400','Unknown'])

plt.show()
```

```python
# Melt 2015 Financial Aid Data
fa = ['no_aptc','yes_aptc','no_csr','yes_csr']
fa15 = pd.melt(df15[fa])

# Convert 2015 Financial Aid to Pivot Table
fa15 = pd.pivot_table(fa15, index = 'variable', values = 'value', aggfunc=np.sum)

# Melt 2016 Financial Aid Data
fa16 = pd.melt(df16[fa])

# Convert 2016 Financial Aid to Pivot Table
fa16 = pd.pivot_table(fa16, index = 'variable', values = 'value', aggfunc=np.sum)

# Plot 2015 Financial Aid Counts
fig, axes = plt.subplots(1,2, figsize=(15,5))

_ = sns.barplot(x=fa15.index, y='value', data=fa15, color='indigo', ax=axes[0])
_ = axes[0].set_xlabel('Financial Assistancet')
_ = axes[0].set_ylabel('Count')
_ = axes[0].set_yticklabels(['{:,}'.format(int(x)) for x in axes[0].get_yticks().tolist()])
_ = axes[0].set_title('2015 Financial Aid Distributions')

_ = sns.barplot(x=fa16.index, y='value', data=fa16, color='indigo', ax=axes[1])
_ = axes[1].set_xlabel('Financial Assistance')
_ = axes[1].set_ylabel('Count')
_ = axes[1].set_yticklabels(['{:,}'.format(int(x)) for x in axes[1].get_yticks().tolist()])
_ = axes[1].set_title('2016 Financial Aid Distributions')

plt.show()
```
